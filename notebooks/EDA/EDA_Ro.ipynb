## Activation Time (KPI)

This is a metric that can be used by the deployment team, as well as the higher ups. It gives a view of kiosks that took a long time to activate after installation, which can point to issues with infrastructure, permitting, technical issues, etc. Using this feature, the team can flag problematic kiosks and get further insight as to factors that contribute to long kiosk activation times.


#### let's find out how long it took kiosks to activate after installation:


```python
df_loc['activation_time'] = (df_loc['activation_complete'] - df_loc['installation_complete'])
long_activation = df_loc[df_loc['activation_time'] >= timedelta(days=60)].sort_values(by='activation_time',ascending=False)
long_activation[['site_id', 'neighborhood_tabulation_area_nta', 'installation_status', 'activation_time']].head(10)

# here is a view of kiosks that took longer than ~2 months to activate after installation. Without further insight as to what is considered "long activation time",
# I simply created a new column that is filterable to find possibly problematic kiosks.
```

## KPI: Average Activation Time

```python
print(f"Average Time From Kiosk Installation to Activation: {long_activation['activation_time'].mean()}")
```

## Energy Efficiency

This is our key metric we want. This will help CityBridge note their energy consumption and how efficient their kiosks are really running. However, given we lack energy usage data, we will have to create an energy consumption proxy using TB uploaded/downloaded, # of sessions, and # of unique clients. This will look like:

<div align='center'>

Energy Efficiency = Data Usage / Energy Consumption Proxy

Where...

Data Usage (GB) = (TB Uploaded + TB Downloaded) * 1024

AND

Energy Consumption Proxy = (Sessions * 0.85) + (Unique Clients * 0.15)

</div>

**NOTE**

- we multiply by 1024 to transform from TB to GB.

- Data Usage is by week, since that is what our dataset is already aggregated by.

- We add sessions by unique clients for an approximation of energy consumption. They represent our user activity load.
    
    - While unique clients may already be represented within, we will add both to try and approximate the higher end of our user activity load.

    - For better fairness, we weigh the factors of our energy consumption proxy by only taking into account a determined percentage of them (thus, 90% of total sessions and 10% of unique users).

```python
# Our Data Usage in GB
data_usage = (df_use['tb_uploaded']+df_use['tb_downloaded'])*1024

# Our weighted user activity load
energy_consumption_proxy = (df_use['sessions']*.85)+(df_use['unique_clients']*.15)

# calculating energy efficiency with the equation we created
df_use['energy_efficiency'] = data_usage/energy_consumption_proxy
```

```python
df_use
```

- Excellent: top 10% (≥ 90th percentile)

- Good: 60–90th percentile

- Fair: 30–60th percentile

- Poor: < 30th percentile

```python
q30 = df_use['energy_efficiency'].quantile(0.30)
q60 = df_use['energy_efficiency'].quantile(0.60)
q90 = df_use['energy_efficiency'].quantile(0.90)

energy_usage_flag = []

for i in df_use['energy_efficiency']:
    if i < q30:
        energy_usage_flag.append('Poor')
    elif q60 > i >= q30:
        energy_usage_flag.append('Fair')
    elif q90 > i >= q60:
        energy_usage_flag.append('Good')
    elif i >= q90:
        energy_usage_flag.append('Excellent')


df_use['energy_usage_flag'] = energy_usage_flag
df_use['energy_usage_flag'].value_counts()
```
