```python
# ==============================================================================
# PART 1: LOAD AND CLEAN DATA
# ==============================================================================
# Thought Process: The first and most critical step is to load the data and
# prepare it for analysis. This involves reading the CSVs, standardizing column
# names for easier access (e.g., lowercase with underscores), and converting
# columns to their correct data types (e.g., dates to datetime objects). A
# try-except block is used to handle the case where the files might not be found,
# allowing the script to run with sample data for demonstration.


# --- Clean Weekly Usage Data ---
# Rename columns to a consistent, code-friendly format
weekly_usage_df.columns = [
    'week_end_date', 'sessions', 'avg_session_length', 'unique_clients',
    'tb_downloaded', 'tb_uploaded', 'cumulative_bandwidth_tb',
    'cumulative_sessions', 'cumulative_subscribers'
]
# Convert date column to datetime objects for time series analysis
weekly_usage_df['week_end_date'] = pd.to_datetime(weekly_usage_df['week_end_date'])
# Set the date as the index to make time-based plotting and resampling easier
weekly_usage_df = weekly_usage_df.set_index('week_end_date').sort_index()

# --- Clean Kiosk Locations Data ---
# Standardize column names
kiosk_locations_df.columns = kiosk_locations_df.columns.str.lower().str.replace(' ', '_').str.replace(r'[^a-zA-Z0-9_]', '', regex=True)
# Convert date columns, coercing errors to 'NaT' (Not a Time) for invalid dates
kiosk_locations_df['installation_complete'] = pd.to_datetime(kiosk_locations_df['installation_complete'], errors='coerce')
kiosk_locations_df['activation_complete'] = pd.to_datetime(kiosk_locations_df['activation_complete'], errors='coerce')


print("\n--- Weekly Usage DataFrame Info ---")
weekly_usage_df.info()
print("\n--- Kiosk Locations DataFrame Info ---")
kiosk_locations_df.info()
```

```python
# ==============================================================================
# PART 2: EDA FOR WEEKLY USAGE DATA
# ==============================================================================
# Thought Process: The goal here is to understand usage patterns over time.
# Line plots are the best tool for visualizing trends. We'll examine sessions
# and unique clients to see overall growth/decline. To check for seasonality,
# we'll aggregate the data by month and use a bar chart for easy comparison.
# Finally, we'll create a 'quality' metric (GB per session) to understand
# user engagement, as required by the business problem.



# --- 2.1: Analyze Trends & Seasonality ---
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10), sharex=True)
weekly_usage_df['sessions'].plot(ax=ax1, title='Weekly Wi-Fi Sessions Over Time', ylabel='Number of Sessions')
weekly_usage_df['unique_clients'].plot(ax=ax2, title='Weekly Unique Clients Over Time', ylabel='Number of Clients', color='orange')
fig.suptitle('Overall Usage Trends', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

```

```python
# Resample data by month and calculate the mean for a stable seasonality view
monthly_avg_sessions = weekly_usage_df['sessions'].resample('M').mean()
# Define month order for correct plotting
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

plt.figure()
sns.barplot(x=monthly_avg_sessions.index.strftime('%B'), y=monthly_avg_sessions.values, order=month_order, palette='viridis')
plt.title('Average Weekly Sessions by Month (Seasonality)')
plt.ylabel('Average Number of Sessions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# --- 2.2: Analyze Usage Quality ---
# Thought Process: To address "efficiency" and "energy waste," we need a proxy for
# usage quality. GB per session is a great metric. It tells us if users are
# engaging in high-bandwidth (valuable) activities.
# Calculate total GB per week (1 TB = 1024 GB)
weekly_usage_df['gb_per_session'] = (weekly_usage_df['tb_downloaded'] + weekly_usage_df['tb_uploaded']) * 1024 / weekly_usage_df['sessions']

plt.figure()
weekly_usage_df['gb_per_session'].plot(title='Average Data Usage per Session Over Time')
plt.ylabel('GB per Session')
plt.tight_layout()
plt.show()

```

```python
# ==============================================================================
# PART 3: EDA FOR KIOSK LOCATIONS DATA
# ==============================================================================
# Thought Process: For the kiosk data, the primary questions are about geography
# and operational status. Bar charts are ideal for this categorical data. We'll
# use `value_counts()` to get the numbers and `seaborn.barplot` to visualize them,
# answering questions about kiosk density and availability.

print("\n--- Part 3: Performing EDA on Kiosk Locations Data ---")

# --- 3.1: Analyze Kiosk Distribution by Borough ---
borough_counts = kiosk_locations_df['borough'].value_counts()
plt.figure()
sns.barplot(x=borough_counts.index, y=borough_counts.values, palette='rocket')
plt.title('Number of LinkNYC Kiosks by Borough')
plt.ylabel('Number of Kiosks')
plt.tight_layout()
plt.show()
```

```python
# --- 3.2: Analyze Kiosk Availability (Installation Status) ---
status_counts = kiosk_locations_df['installation_status'].value_counts()
plt.figure()
sns.barplot(x=status_counts.index, y=status_counts.values, palette='mako')
plt.title('Kiosk Installation Status')
plt.ylabel('Number of Kiosks')
plt.tight_layout()
plt.show()
```
