## Activation Time

This is a metric that can be used by the deployment team, as well as the higher ups. It gives a view of kiosks that took a long time to activate after installation, which can point to issues with infrastructure, permitting, technical issues, etc. Using this feature, the team can flag problematic kiosks and get further insight as to factors that contribute to long kiosk activation times.


#### let's find out how long it took kiosks to activate after installation:


```python
df_loc['activation_time'] = (df_loc['activation_complete'] - df_loc['installation_complete'])
long_activation = df_loc[df_loc['activation_time'] >= timedelta(days=60)].sort_values(by='activation_time',ascending=False)
long_activation[['site_id', 'neighborhood_tabulation_area_nta', 'installation_status', 'activation_time']].head(10)

# here is a view of kiosks that took longer than ~2 months to activate after installation. Without further insight as to what is considered "long activation time",
# I simply created a new column that is filterable to find possibly problematic kiosks.
```

***

## Kiosk Age

This metric can be useful for finding the oldest running kiosks, which can be used by the maintenance team to prioritize which kiosks need attention the most. Going forward, a new column can also be created that would mark the latest maintenence date.

#### Let's explore the oldest live kiosks:

```python
df_loc['kiosk_age'] = datetime.now() - df_loc['activation_complete']
oldest_kiosks = df_loc[df_loc['kiosk_age']>= timedelta(weeks=520)].sort_values(by='kiosk_age', ascending=False)
oldest_kiosks[['site_id', 'neighborhood_tabulation_area_nta', 'installation_status', 'kiosk_age']]

# Another simple view showing the oldest activated kiosks. The 4 displayed kiosks have been activated for longer than 10 years and are still live. These should be considered for
# repairs/reviews, as old equipment becomes more prone to technical issues. This new column can also be custom-filtered.
```

## Weekly Session Change

This metric will be useful for doing a weekly track of session changes, and can be supported by other data sources to find factors contributing to session changes.

#### Let's find the difference in percentages of sessions per week:

```python
# data is already per ending week:
df_use['weekly_session_change_pct'] = round(((df_use['sessions'].diff())/df_use['sessions'].shift(1))*100,2)
df_use[['week_end_date', 'sessions', 'weekly_session_change_pct']]
```

***

## Average Time per Week

This can help support our weekly session change column, by showing the actually length of average session usage per week. I also included an overall calculation, to show how much LinkNYC's usage has increased since its inception.

#### Our data is already aggregated by end of week, so let's calculate the avg session length:

```python
# let's create new column: avg_session_length_seconds as integer to be able to use for calculations
# this will be used when aggregating per kiosk for energy consumption (alongside a proxy of upload/download amounts)
df_use['avg_session_length_seconds'] = pd.to_datetime(df_use['avg_session_length'])
df_use['avg_session_length_seconds'] = ((df_use['avg_session_length_seconds'].dt.hour)*2400) + ((df_use['avg_session_length_seconds'].dt.minute)*60) + df_use['avg_session_length_seconds'].dt.second
df_use['avg_session_length_minutes'] = round(df_use['avg_session_length_seconds']/60,2)

# since our data is already aggregated by ending week, we simply display the two columns:
session_usage_minutes = df_use[['week_end_date', 'avg_session_length_minutes']]
session_usage_minutes
```

```python
print(f"Session Length Change Overall: {round(((session_usage_minutes['avg_session_length_minutes'][299]-session_usage_minutes['avg_session_length_minutes'][0])/session_usage_minutes['avg_session_length_minutes'][0])*100,2)}%")
```
